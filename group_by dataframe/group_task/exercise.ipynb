```python
import numpy as np
import pandas as pd
```

```python
dl= pd.read_csv("deliveries.csv")
dl.head(5)
```

```python
#top 10 batsman run

dl.groupby("batsman")["batsman_runs"].sum().sort_values(ascending=False).head(10)
```

```python
#total no. of 6

six=dl[dl["batsman_runs"]==6]
six.groupby("batsman")["batsman_runs"].count().sort_values(ascending=False).head(3)
```

```python
#find batsamn who last 5 over me jaada 4 aur 6 lagaya hai

lst=dl[dl["over"]>15]
temp=lst[(lst["batsman_runs"]==4) | (lst["batsman_runs"]==6)]
temp.groupby("batsman")["batsman"].count().sort_values(ascending=False)
```

```python
#vk runs against all team

vk=dl[dl["batsman"]=="V Kohli"]
vk.groupby("bowling_team")["batsman_runs"].sum().reset_index()
```
