```python
import numpy as np
import pandas as pd
```

```python
#multiindex 
multi_index=[("cse",2019),("cse",2020),("cse",2021),("cse",2022),("ece",2019),("ece",2020),("ece",2021),("ece",2022)]
#pd.MultiIndex.from_tuples
#pd.MultiIndex.from_product
tuple_ji=pd.MultiIndex.from_tuples(multi_index)
product_ji=pd.MultiIndex.from_product([["cse","ece"],[2019,2020,2021,2022]])
product_ji
```

```python
#level inside multiindex
```

```python
#creating a series with multiindex 
s=pd.Series([1,2,3,4,5,6,7,8],index=tuple_ji)
s
```

```python
#fetch the items 
s["ece"]
```

```python
#unstack

temp=s.unstack()
temp
```
